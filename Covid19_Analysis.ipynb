```python
vaccine_df =  pd.read_csv("covid_vaccine_statewise.csv")
```

```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"],inplace=True,axis = 1)
```

```python
covid_df.isnull().sum()
```

```python
covid_df.dropna(inplace=True)
```

```python
covid_df.head()
```

```python
covid_df["Date"] = pd.to_datetime(covid_df["Date"], format = "%Y-%m-%d")
covid_df["Confirmed"] = covid_df["Confirmed"].astype(int)
covid_df["Deaths"] = covid_df["Deaths"].astype(int)
covid_df["Cured"] = covid_df["Cured"].astype(int)
```

```python
covid_df.head()
```

```python
#Active Cases

covid_df["Active_Cases"] = covid_df["Confirmed"] - covid_df["Cured"] - covid_df["Deaths"]
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df,values = ["Confirmed","Deaths","Cured"],index = "State/UnionTerritory",aggfunc = "max")
```

```python
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Death Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by = "Confirmed",ascending = False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
# Top 10 actiive cases states

top_10_active_cases = covid_df.groupby("State/UnionTerritory").max()[["Active_Cases",'Date']].sort_values(by = ["Active_Cases"],ascending = False).head(10).reset_index()
```

```python
fig = plt.figure(figsize = (16,9))
```

```python
plt.title("Top 10 States with most Active Cases in India",size = 25)
```

```python
ax = sns.barplot(data = top_10_active_cases.iloc[:10],y = "Active_Cases",x = "State/UnionTerritory",linewidth = 2, edgecolor = "red")
```

```python
# Top 10 actiive cases states

top_10_active_cases = covid_df.groupby("State/UnionTerritory").max()[["Active_Cases",'Date']].sort_values(by = ["Active_Cases"],ascending = False).head(10).reset_index()

fig = plt.figure(figsize = (16,9))

plt.title("Top 10 States with most Active Cases in India",size = 25)

ax = sns.barplot(data = top_10_active_cases.iloc[:10],y = "Active_Cases",x = "State/UnionTerritory",linewidth = 2, edgecolor = "red")

plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
# Top states with highest deaths

top_10_deaths = covid_df.groupby("State/UnionTerritory").max()[["Deaths",'Date']].sort_values(by = ["Deaths"],ascending = False).head(10).reset_index()

fig = plt.figure(figsize = (18,5))

plt.title("Top 10 States with most Deaths in India",size = 25)

ax = sns.barplot(data = top_10_deaths.iloc[:12],y = "Deaths",x = "State/UnionTerritory",linewidth = 2, edgecolor = "black")
xlabel = plt.xlabel("States")
ylabel = plt.ylabel("Total Deaths")
plt.show()
```

```python
# Growth trend

fig = plt.figure(figsize = (12,6))

ax = sns.lineplot(data = covid_df[covid_df["State/UnionTerritory"].isin(["Maharashtra","Karnataka","Kerala","Tamil Nadu", "Uttar Pradesh"])],x = "Date", y = "Active_Cases", hue = "State/UnionTerritory")

ax.set_title("Top 5 Affected States in India",size = 16)
```

```python
# Remove rows where state is India

vaccine = vaccine_df[vaccine_df["State"]!="India"]
vaccine
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated":"Total"},inplace=True)
vaccine.head()
```

```python
# most vaccinated states

max_vac = vaccine.groupby("State")["Total"].sum().to_frame()
max_vac = max_vac.sort_values(by="Total",ascending=False)[:5]
max_vac
```

```python
fig = plt.figure(figsize = (16,9))

plt.title("Top 5 States with most Vaccination in India",size = 25)

ax = sns.barplot(data = max_vac.iloc[:10],y = max_vac.Total ,x = max_vac.index ,linewidth = 2, edgecolor = "red")

plt.xlabel("States")
plt.ylabel("Total Vaccination")
plt.show()
```
